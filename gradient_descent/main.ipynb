```python
import pandas as pd 

df = pd.read_csv("../data/clean_weather.csv")
df = df.ffill()
```

```python
df.head()
```

```python
df.plot.scatter("tmax", "tmax_tomorrow")

import matplotlib.pyplot as plt 

plt.plot([30, 120], [30, 120], "red")
```

```python
from sklearn.linear_model import LinearRegression 

lr = LinearRegression() 

lr.fit(df[["tmax"]], df["tmax_tomorrow"])
```

```python
df.plot.scatter("tmax", "tmax_tomorrow") 
plt.plot(df["tmax"], lr.predict(df[["tmax"]]), "purple")
```

```python
print(f"Weight: {lr.coef_[0]:.2f}")
print(f"Bias: {lr.intercept_:.2f}")
```
